```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the CSV file
df = pd.read_csv("../csvs/batch_results.csv")
```

```python
df.head(15)
```

# Distributions

## Area

```python
# Max Contour Area distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['maxContourArea'], bins=30, color="#03051A", alpha=1)
plt.title("Max Contour Area Distribution")
plt.xlabel("Max Contour Area")
plt.ylabel("Frequency")
plt.show()
```

```python
# Max Contour Area distribution by gesture
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='maxContourArea', hue='gesture', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max Contour Area by Gesture')
plt.xlabel('Max Contour Area')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

## Used Memory

```python
# Max Memory distribution
sns.histplot(df['usedMemoryMB'], bins=30, color="#03051A", alpha=1)
plt.title("Max Memory Distribution")
plt.xlabel("Max Memory (MB)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Max usedMemoryMB distribution by fps
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='usedMemoryMB', hue='gesture', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max FPS by usedMemoryMB')
plt.xlabel('Max usedMemoryMB')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

## CPU Load

```python
# CPU distribution
sns.histplot(df['cpuLoad'], bins=30, color="#03051A", alpha=1)
plt.title("CPU Load Distribution")
plt.xlabel("CPU Load (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Max cpuLoad distribution by fps
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='cpuLoad', hue='gesture', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max FPS by cpuLoad')
plt.xlabel('Max cpuLoad')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
sns.lmplot(
    data=df, x="maxContourArea", y="avgAngle", col="gesture", hue="fingers",
    col_wrap=2, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)

# this plot shows a linear relationship between maxContourArea and centerY for each gesture
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='gesture', y='maxContourArea', palette='Set2', inner='quartile')
plt.title('Violin Plot of Max Contour Area by Gesture')
plt.xlabel('Gesture')
plt.ylabel('Max Contour Area')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# FPS
axs[0].plot(df['frame'], df['fps'], color='tab:blue')
axs[0].set_ylabel('FPS')
axs[0].set_title('FPS over Frames')
axs[0].grid(True)

# Memory
axs[1].plot(df['frame'], df['usedMemoryMB'], color='tab:green')
axs[1].set_ylabel('Used Memory (MB)')
axs[1].set_title('Memory Usage over Frames')
axs[1].grid(True)

# CPU Load
axs[2].plot(df['frame'], df['cpuLoad'], color='tab:red')
axs[2].set_ylabel('CPU Load')
axs[2].set_xlabel('Frame')
axs[2].set_title('CPU Load over Frames')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Distribuição de gestos - gráfico de barras
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gesture', palette='Set3', order=df['gesture'].value_counts().index)
plt.title('Distribuição de Gestos')
plt.xlabel('Gestos')
plt.ylabel('Contagem')
plt.xticks(rotation=30)
plt.show()

# Distribuição de gestos - gráfico de pizza
plt.figure(figsize=(6, 6))
df['gesture'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Distribuição de Gestos (Pizza)')
plt.ylabel('')
plt.show()
```

```python
# Scatter plot de centerX vs centerY
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='centerX', y='centerY', hue='gesture', palette='Set2', alpha=0.7)
plt.title('Distribuição da Mão na Tela (centerX vs centerY)')
plt.xlabel('centerX')
plt.ylabel('centerY')
plt.legend(title='Gesture', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# Linha de centerX e centerY ao longo do frame
plt.figure(figsize=(14, 5))
plt.plot(df['frame'], df['centerX'], label='centerX', color='tab:blue')
plt.plot(df['frame'], df['centerY'], label='centerY', color='tab:orange')
plt.title('Deslocamento Horizontal e Vertical ao Longo dos Frames')
plt.xlabel('Frame')
plt.ylabel('Posição')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Linha de fingers ao longo do frame
plt.figure(figsize=(14, 5))
plt.plot(df['frame'], df['fingers'], marker='o', linestyle='-', color='tab:purple', alpha=0.7)
plt.title('Quantidade de Dedos Detectados ao Longo dos Frames')
plt.xlabel('Frame')
plt.ylabel('Fingers')
plt.grid(True)
plt.show()

# Bar chart de gesture (frequência de cada gesto)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gesture', palette='Set2', order=df['gesture'].value_counts().index)
plt.title('Frequência de Cada Gesto Detectado')
plt.xlabel('Gesto')
plt.ylabel('Contagem')
plt.xticks(rotation=30)
plt.show()

# Scatter fingers vs avgAngle
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='fingers', y='avgAngle', hue='gesture', palette='Set1', alpha=0.8)
plt.title('Relação entre Quantidade de Dedos e Ângulo Médio')
plt.xlabel('Fingers')
plt.ylabel('Avg Angle')
plt.legend(title='Gesture', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Linha de maxContourArea ao longo do frame
plt.figure(figsize=(14, 5))
plt.plot(df['frame'], df['maxContourArea'], color='tab:blue')
plt.title('Variação da Área Máxima do Contorno ao Longo dos Frames')
plt.xlabel('Frame')
plt.ylabel('Max Contour Area')
plt.grid(True)
plt.show()

# Scatter plot de maxContourArea vs convexDefects
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='maxContourArea', y='convexDefects', hue='gesture', palette='Set1', alpha=0.8)
plt.title('Relação entre Área Máxima do Contorno e Defeitos Convexos')
plt.xlabel('Max Contour Area')
plt.ylabel('Convex Defects')
plt.legend(title='Gesture', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Linha de fps ao longo do frame para identificar quedas de performance
plt.figure(figsize=(14, 5))
plt.plot(df['frame'], df['fps'], color='tab:blue')
plt.title('Variação do FPS ao Longo dos Frames')
plt.xlabel('Frame')
plt.ylabel('FPS')
plt.grid(True)
plt.show()

# Linha de usedMemoryMB e cpuLoad ao longo do frame para monitoramento de recursos
plt.figure(figsize=(14, 5))
plt.plot(df['frame'], df['usedMemoryMB'], label='Used Memory (MB)', color='tab:green')
plt.plot(df['frame'], df['cpuLoad'] * 100, label='CPU Load (%)', color='tab:red')  # cpuLoad em %
plt.title('Uso de Memória e CPU ao Longo dos Frames')
plt.xlabel('Frame')
plt.ylabel('Uso de Recursos')
plt.legend()
plt.grid(True)
plt.show()

# Scatter plot de fps vs maxContourArea para investigar impacto de mãos grandes no FPS
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='maxContourArea', y='fps', hue='gesture', palette='Set1', alpha=0.8)
plt.title('FPS vs Max Contour Area')
plt.xlabel('Max Contour Area')
plt.ylabel('FPS')
plt.legend(title='Gesture', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
sns.set_theme()
g = sns.JointGrid(data=df, x="fps", y="maxContourArea", space=0)
g.plot_joint(
    sns.kdeplot,
    fill=True,
    clip=((df['fps'].min(), df['fps'].max()), (df['maxContourArea'].min(), df['maxContourArea'].max())),
    thresh=0, levels=100, cmap="rocket"
)
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
plt.show()
```

```python
sns.set_theme()
g = sns.JointGrid(data=df, x="centerY", y="centerX", space=0)
g.plot_joint(
    sns.kdeplot,
    fill=True,
    clip=((df['centerY'].min(), df['centerY'].max()), (df['centerX'].min(), df['centerX'].max())),
    thresh=0, levels=100, cmap="rocket"
)
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
plt.show()
```

```python

# Heatmap centerX vs centerY (mapa de calor do movimento da mão)
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='centerX', y='centerY', fill=True, cmap='mako', thresh=0.005)
plt.title('Mapa de Calor do Movimento da Mão (centerX vs centerY)')
plt.xlabel('centerX')
plt.ylabel('centerY')
plt.show()

# Scatter fingers vs convexDefects (validar consistência do algoritmo de contagem de dedos)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='fingers', y='convexDefects', hue='gesture', palette='Set1', alpha=0.8)
plt.title('Fingers vs Convex Defects')
plt.xlabel('Fingers')
plt.ylabel('Convex Defects')
plt.legend(title='Gesture', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# Scatter avgAngle vs gesture (entender variação de ângulo por tipo de gesto)
plt.figure(figsize=(10, 6))
sns.stripplot(data=df, x='gesture', y='avgAngle', palette='Set2', alpha=0.7, jitter=True)
plt.title('Variação de Ângulo Médio por Tipo de Gesto')
plt.xlabel('Gesture')
plt.ylabel('Avg Angle')
plt.xticks(rotation=30)
plt.grid(True, axis='y')
plt.show()
```
