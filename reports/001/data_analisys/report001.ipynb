```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the CSV file
df = pd.read_csv("../csvs/performance_496.csv")
```

```python
df.head(15)
```

# Distributions

## Area

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['maxContourArea'], bins=30, stat='density', color="#03051A", alpha=1)
mu, std = df['maxContourArea'].mean(), df['maxContourArea'].std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
from scipy.stats import norm
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2, label='Gaussian Fit')
plt.title("Max Contour Area Distribution with Gaussian Fit")
plt.xlabel("Max Contour Area")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Max Contour Area distribution by gesture
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='maxContourArea', hue='gesture', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max Contour Area by Gesture')
plt.xlabel('Max Contour Area')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Max Contour Area by FPS
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='maxContourArea', hue='fps', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max Contour Area by FPS')
plt.xlabel('Max Contour Area')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

## FPS

```python
# Max Contour FPS distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['fps'], bins=30, color="#03051A", alpha=1)
plt.title("Max FPS Distribution")
plt.xlabel("Max FPS")
plt.ylabel("Frequency")
plt.show()
```

```python

# Max FPS distribution by gesture
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='fps', hue='gesture', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max FPS by Gesture')
plt.xlabel('Max FPS')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

## Used Memory

```python
# Max Memory distribution
sns.histplot(df['usedMemoryMB'], bins=30, stat='density', color="#03051A", alpha=1)
mu, std = df['usedMemoryMB'].mean(), df['usedMemoryMB'].std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
from scipy.stats import norm
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2, label='Gaussian Fit')
plt.title("Max Memory Distribution with Gaussian Fit")
plt.xlabel("Max Memory (MB)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Max usedMemoryMB distribution by fps
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='usedMemoryMB', hue='gesture', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max FPS by usedMemoryMB')
plt.xlabel('Max usedMemoryMB')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

## CPU Load

```python
# CPU distribution
from scipy.stats import norm
y = df['cpuLoad']
# Fit a normal distribution to the data
mu, std = norm.fit(y)

# Plot the histogram and the PDF
plt.figure(figsize=(10, 6))
sns.histplot(y, bins=30, kde=False, stat='density', color="#03051A", alpha=1, label='FPS Histogram')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2, label='Gaussian Fit')

plt.title("FPS Distribution with Gaussian Fit")
plt.xlabel("FPS")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Max cpuLoad distribution by fps
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='cpuLoad', hue='gesture', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max FPS by cpuLoad')
plt.xlabel('Max cpuLoad')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

## Gesture

```python
sns.histplot(df['gesture'], bins=30, color="#03051A", alpha=1, stat='density')
from scipy.stats import norm

# Convert gestures to categorical codes for fitting
gesture_codes = df['gesture'].astype('category').cat.codes
mu, std = gesture_codes.mean(), gesture_codes.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2, label='Gaussian Fit')
plt.title("Gesture Distribution with Gaussian Fit")
plt.xlabel("Gesture")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Max gesture distribution by fps
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='gesture', hue='fps', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max FPS by gesture')
plt.xlabel('Max gesture')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
sns.set_theme()
g = sns.JointGrid(data=df, x="fps", y="maxContourArea", space=0)
g.plot_joint(
    sns.kdeplot,
    fill=True,
    clip=((df['fps'].min(), df['fps'].max()), (df['maxContourArea'].min(), df['maxContourArea'].max())),
    thresh=0, levels=100, cmap="rocket"
)
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
plt.show()
```

```python
# Comparações múltiplas entre usedMemoryMB, cpuLoad, fps, maxContourArea e gesture

# Boxplot de cada métrica por gesture
fig, axs = plt.subplots(2, 2, figsize=(16, 10))
sns.boxplot(data=df, x='gesture', y='usedMemoryMB', ax=axs[0,0], palette='Set3')
axs[0,0].set_title('Used Memory (MB) por Gesto')
axs[0,0].set_xticklabels(axs[0,0].get_xticklabels(), rotation=30)

sns.boxplot(data=df, x='gesture', y='cpuLoad', ax=axs[0,1], palette='Set3')
axs[0,1].set_title('CPU Load por Gesto')
axs[0,1].set_xticklabels(axs[0,1].get_xticklabels(), rotation=30)

sns.boxplot(data=df, x='gesture', y='fps', ax=axs[1,0], palette='Set3')
axs[1,0].set_title('FPS por Gesto')
axs[1,0].set_xticklabels(axs[1,0].get_xticklabels(), rotation=30)

sns.boxplot(data=df, x='gesture', y='maxContourArea', ax=axs[1,1], palette='Set3')
axs[1,1].set_title('Max Contour Area por Gesto')
axs[1,1].set_xticklabels(axs[1,1].get_xticklabels(), rotation=30)

plt.tight_layout()
plt.show()

# Scatterplot matrix customizada
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='usedMemoryMB', y='fps', hue='gesture', palette='Set1', alpha=0.7)
plt.title('FPS vs Used Memory por Gesto')
plt.xlabel('Used Memory (MB)')
plt.ylabel('FPS')
plt.legend(title='Gesture', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='cpuLoad', y='fps', hue='gesture', palette='Set2', alpha=0.7)
plt.title('FPS vs CPU Load por Gesto')
plt.xlabel('CPU Load')
plt.ylabel('FPS')
plt.legend(title='Gesture', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='maxContourArea', y='usedMemoryMB', hue='gesture', palette='Set3', alpha=0.7)
plt.title('Used Memory vs Max Contour Area por Gesto')
plt.xlabel('Max Contour Area')
plt.ylabel('Used Memory (MB)')
plt.legend(title='Gesture', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Heatmap de média de convexDefects por gesture e quantidade de dedos
pivot_defects = df.pivot_table(index='gesture', columns='fps', values='convexDefects', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_defects, annot=False, fmt=".1f", cmap='YlOrBr')
plt.title('Média de Convex Defects por Gesto e Quantidade de Dedos')
plt.xlabel('fps')
plt.ylabel('Gesture')
plt.show()

# Heatmap de média de usedMemoryMB por gesture e quantidade de dedos
pivot_mem = df.pivot_table(index='gesture', columns='fps', values='usedMemoryMB', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_mem, annot=False, fmt=".1f", cmap='Greens')
plt.title('Média de Used Memory (MB) por Gesto e Quantidade de Dedos')
plt.xlabel('fps')
plt.ylabel('Gesture')
plt.show()

pivot_cpu = df.pivot_table(index='gesture', columns='fps', values='cpuLoad', aggfunc='mean')
# Heatmap de média de cpuLoad por gesture e quantidade de dedos
pivot_cpu_pct = (pivot_cpu * 100).round(2)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_cpu_pct, annot=False, fmt=".2f", cmap='Blues')
plt.title('Média de CPU Load por Gesto e Quantidade de Dedos')
plt.xlabel('fps')
plt.ylabel('Gesture')
plt.show()
```
